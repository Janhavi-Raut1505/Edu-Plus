```python
# Q27. Write a program to check whether a number is palindrome or not using a while loop.

num = int(input("Enter the number:"))
original = num
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
if original == reverse:
    print("It's a palindrome")
else:
    print("It's not a palindrome")
```

```python
# Q34. write a program to check whether a number is armstrong or not using while loop.

# an armstrong number is a number in which the sum of the cubes of its digits is equal to the original number.
# example:
# 153 = 1^(3) + 5^(3) + 3^(3)
# 153 = 1 + 125 + 27
# 153 = 153
# So, 153 is an aemstrong number.

num = int(input("Enter the number:"))
temp = num  # stores original number for checking later
total = 0   # variable to store sum

while temp > 0:  # loop runs until number becomes 0
    digit = temp % 10  # get last digit of the number
    total = total + (digit * digit * digit)  # adds cube of each digit
    temp = temp // 10  # remove last digit from the number
    
if total == num:  #checks whether sum is equal to original number
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
# Q35. Write a program to print Fibonacci series up to n terms using a while loop.
# The Fibonacci series is a simple pattern of numbers where you get the next number by adding the last two numbers together.
# It starts with 0 and 1.

series = int(input("Enter number:"))
a = 0
b = 1
count = 1
while count <= series:
    print(a)
    c = a + b
    a = b
    b = c
    count += 1
```

```python
# Q37. Write a program to find the least common multiple (LCM) of two numbers using a while loop.

a = int(input("Enter first number:"))
b = int(input("Enter second number:"))
max_num = max(a, b)
while True:
    if max_num % a == 0 and max_num % b == 0:
        print("LCM:", max_num)
        break
    max_num += 1
```

```python
# Q39. Write a program to count positive, negative, and zero values from user inputs using a while loop.

positive = 0
negative = 0
zero = 0
n = int(input("How many numbers:"))
i = 1
while i <= n:
    num = int(input("Enter number:"))
    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1
    i += 1
print("Positive:", positive)
print("Negative:", negative)
print("Zero:", zero)
```

```python
# Q49. Write a program to check whether a number is perfect or not using a while loop.

num = int(input("Enter number: "))

sum = 0
i = 1

while i < num:
    if num % i == 0:
        sum += i
    i += 1

if sum == num:
    print("Perfect Number")
else:
    print("Not Perfect Number")
```
