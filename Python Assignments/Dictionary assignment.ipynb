```python
# 12. create a dictionary of employees with their departments.

employees = {"Amit": "HR", "Neha": "IT", "Rohan": "Finance", "Priya": "Marketing"}
print(employees)
```

```python
# 13. Access the department of a specific employee.

print("Priya's Department:", employees["Priya"])
```

```python
# 14. Add a new employee and department.

employees["Karan"] = "Sales"
print(employees)
```

```python
# 15. Change the department of an existing employee.

employees["Amit"] = "Admin"
print("Changed department of Amit:", employees)
```

```python
# 16. Remove an employee from the dictionary.

employees.pop("Neha")
print("Aftering removing the employee:", employees)
```

```python
# 17. Sort the employee dictionary by employee names in ascending order.

sorted_by_keys = dict(sorted(employees.items()))
print("Sorted names of the employees in ascending order:", sorted_by_keys)
```

```python
# 18. Sort the employee dictionary by department names in ascending order.

sorted_by_values_asc = dict(sorted(employees.items(), key=lambda x: x[1]))
print("Sorted departments of the employees in ascending order:", sorted_by_values_asc)
```

```python
# 19. Sort the employee dictionary by names in descending order.

sorted_names_desc = dict(sorted(employees.items(), key=lambda x: x[1], reverse=True))
print("Sorted the employees by names in descending order:", sorted_names_desc)
```

```python
# 20.  Sort the employee dictionary by department in descending order.

sorted_dept_desc = dict(sorted(employees.items(), key=lambda x: x[1], reverse=True))
print("Sorted employees by department in descending order:", sorted_dept_desc)
```

```python
# 23. Create a nested dictionary for 3 students with age, grade, and department.

student_info = {
    "Janhavi": {"age": 19, "grade": "A", "department": "BCA"},
    "Saaee": {"age": 20, "grade": "B", "department": "BSc"},
    "Raj": {"age": 21, "grade": "A+", "department": "IT"}
}
print(student_info)
```

```python
# 29. Copy a dictionary into a new dictionary and modify the copy.

student = {"name": "Janhavi", "age": 21, "course": "BCA"}
print(student)
new_student = student.copy()
new_student['name'] = "Sharda"
new_student['age'] = 22
new_student['course'] = "MCA"
print("Modify the copy:", new_student)
```
