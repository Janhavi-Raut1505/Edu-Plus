```python
# Q2. Build a project tracker where nested functions update project completion status dynamically. 

def project_tracker(project):
    completion = 0  # initilize completion
    def update(status):
        nonlocal completion  # refers to variable from the outer function (accessing enclosing variable)
        completion += status  # modifies the enclosing variable (updates completion)
        print(project, "Completion:", completion, "%")
    return update
p = project_tracker("Website")
p(20)
p(30)
```

```python
# Q3. Develop a Python function that maintains cumulative monthly expenses without using global variables.

# outer function
def expenses_tracker():
    # enclosing variable (storing total expense)
    total = 0

    # inner function
    def add_expense(amount):
        nonlocal total  # updating outer function
        total += amount  # adding expense
        print("Total Expense=", total)
    return add_expense  # returning function
expenses = expenses_tracker()  # creating tracker
expenses(2000)  # adding expenses
expenses(3000)
```

```python
# Q4. Create a ticket booking counter using closures for seat allocation tracking.

# Outer function
def ticket_counter(total_seats):
    
    # Total seats available
    seats = total_seats

    # Inner function
    def book_ticket():

        # Using outer variable
        nonlocal seats

        # Checking seats
        if seats > 0:

            # Booking one seat
            seats -= 1

            # Printing booked message
            print("Seat Booked Successfully")

            # Printing remaining seats separately
            print("Remaining Seats =", seats)

        else:

            # If no seats available
            print("No Seats Available")

    # Returning inner function
    return book_ticket


# Taking input from user
total = int(input("Enter Total Seats: "))

# Creating counter
counter = ticket_counter(total)

# Taking booking attempts
times = int(input("Enter Number of Bookings: "))

# Loop for bookings
for i in range(times):

    # Calling booking function
    counter()
```

```python
# Q5. Build a bank transaction logger where each function maintains independent transaction histories.

# Outer function
def bank_account():
    # Empty transaction history
    history = []

    # Inner function
    def transaction(amount):
        # Adding transaction to list
        history.append(amount)
        print("Transaction History =", history)

    # Returning function
    return transaction

# Creating two independent accounts
acc1 = bank_account()
acc2 = bank_account()

# Transactions for account 1
acc1(1000)
acc1(-500)

# Transaction for account 2
acc2(3000)
```

```python
# Q7. Create a recursive user-defined function to flatten a nested list structure.

# recursive function
def flatten(lst):
    result = []  # empty result list
    for item in lst:
        if type(item) == list:
            result.extend(flatten(item))  # recursive call
        else:
            result.append(item)  # adding normal element
    return result
print(flatten([1, [2, 3], [4, [5, 6]]]))
```

```python
# Q12. Create a custom text encryption-decryption utility using functions.

def encrypt(text):
    result = ""
    for ch in text:
        result += chr(ord(ch) + 1)
    return result

def decrypt(text):
    result = ""
    for ch in text:
        result += chr(ord(ch) - 1)
    return result

msg = "HELLO"

encrypted = encrypt(msg)
print("Encrypted:", encrypted)

decrypted = decrypt(encrypted)
print("Decrypted:", decrypted)
```

```python
# Q14. Develop a Python iterator for Fibonacci numbers with custom start and stop ranges.

class Fibonacci:
    def __init__(self, start, stop):
        self.a = 0
        self.b = 1
        self.stop = stop

    def __iter__(self):
        return self

    def __next__(self):
        if self.a > self.stop:
            raise StopIteration

        value = self.a
        self.a, self.b = self.b, self.a + self.b
        return value

for num in Fibonacci(0, 50):
    print(num, end=" ")
```

```python
# Q15. Create an iterator that cycles infinitely through weekdays.

class Weekdays:
    def __init__(self):
        self.days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        day = self.days[self.index]
        self.index = (self.index + 1) % len(self.days)
        return day

week = Weekdays()

for i in range(10):
    print(next(week))
```

```python
# Q16. Design an iterator that traverses nested dictionaries level by level.

class DictIterator:
    def __init__(self, data):
        self.items = []

        for key, value in data.items():
            self.items.append((key, value))

        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.items):
            raise StopIteration

        item = self.items[self.index]
        self.index += 1
        return item

data = {"Name": "Janhavi", "Age": 20, "City": "Pune"}

for i in DictIterator(data):
    print(i)
```

```python
# Q17. Build a custom iterator to iterate through alternate elements of a list.

class Alternate:
    def __init__(self, data):
        self.data = data
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.data):
            raise StopIteration

        value = self.data[self.index]
        self.index += 2
        return value

numbers = [10, 20, 30, 40, 50, 60]

for n in Alternate(numbers):
    print(n)
```

```python
# Q18. Create an iterator that yields cumulative sums from a dataset.

class CumulativeSum:
    def __init__(self, numbers):
        self.numbers = numbers
        self.index = 0
        self.total = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.numbers):
            raise StopIteration

        self.total += self.numbers[self.index]
        self.index += 1
        return self.total

data = [10, 20, 30, 40]

for i in CumulativeSum(data):
    print(i)
```

```python
# Q19. Design a reverse iterator for a sentence word-by-word traversal.

class ReverseSentence:
    def __init__(self, sentence):
        self.words = sentence.split()
        self.index = len(self.words) - 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index < 0:
            raise StopIteration

        word = self.words[self.index]
        self.index -= 1
        return word

s = ReverseSentence("I Love Python")

for word in s:
    print(word)
```

```python
# Q20. Build an iterator that reads and processes CSV rows one at a time.

import csv

class CSVReader:
    def __init__(self, filename):
        self.file = open(filename, "r", encoding="utf-8-sig")
        self.reader = csv.reader(self.file)

    def __iter__(self):
        return self

    def __next__(self):
        return next(self.reader)

obj = CSVReader("Data_py.csv")

for row in obj:
    print(row)
```

```python
# Q23. Develop a closure-based OTP verification system with limited attempts.

def otp_system():
    otp = "1234"
    attempts = 3

    def verify():
        nonlocal attempts

        while attempts > 0:
            user_otp = input("Enter OTP: ")

            if user_otp == otp:
                print("OTP Verified Successfully")
                return
            else:
                attempts -= 1
                print("Wrong OTP")
                print("Remaining Attempts:", attempts)

        print("No Attempts Left")

    return verify

check = otp_system()
check()
```

```python
# Q28. Develop a generator that yields Pascal Triangle rows dynamically.

def pascal(n):
    row = [1]
    for i in range(n):
        yield row
        row = [1] + [row[j] + row[j+1] for j in range(len(row)-1)] + [1]

for r in pascal(5):
    print(r)
```

```python
# Q30. Build a generator that produces rolling averages from incoming numbers.

def rolling_average(numbers):
    total = 0

    for i, num in enumerate(numbers, 1):
        total += num
        yield total / i

for avg in rolling_average([10,20,30]):
    print(avg)
```

```python
# Q31. Design a generator that yields all permutations of a string one at a time.

from itertools import permutations

def generate(word):
    for p in permutations(word):
        yield ''.join(p)

for i in generate("ABC"):
    print(i)
```

```python
# Q37. Create a decorator that caches function outputs for repeated inputs.

def cache(func):
    memory = {}

    def wrapper(n):
        if n in memory:
            return memory[n]

        memory[n] = func(n)
        return memory[n]

    return wrapper

@cache
def square(n):
    return n*n

print(square(5))
```
