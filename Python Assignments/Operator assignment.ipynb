```python
# 1. Write a Python program to add two numbers entered by the user.

a = 10
b = 20
sum = a + b

print("Addition (a + b)", sum)
```

```python
# 2. Write a Python program to calculate the area of a rectangle using multiplication (*).

length = 15
width = 8
area = length * width

print("The area of a rectangle:", area)
```

```python
# 3. Write a Python program to find the remainder when one number is divided by another using modulus (%).

a = 25
b = 4
mod_result = a % b

print("Remainder:", mod_result)
```

```python
# 4. Write a Python program to calculate the square and cube of a given number using the exponent (**) operator.

a = 5
square = a ** 2
cube = a ** 3

print("Square of 5:", square)
print("Cube of 5:", cube)
```

```python
# 5. Write a Python program to convert total seconds into minutes and seconds using division and modulus.

sec = 250
minutes = sec // 60

remaining_sec = sec % 60
print("Minutes:", minutes)
print("Seconds:", sec)
```

```python
# 6. Write a Python program to find the average of 5 numbers using arithmetic operators.

a = 10
b = 20
c = 30
d = 40
e = 50
avg = (a + b + c + d + e) / 5
print("Average of the 5 numbers:", avg)
```

```python
# 7. Write a Python program to swap two numbers without using a third variable (use + and - operators).

a = 15
b = 25
a = a + b
b = a - b
print("After swapping:", "a:", a, "b:", b)
```

```python
# 13. Write a Python program to check if the entered password length is less than or equal to 8.

password = "admin123"
print("The entered password length is less than or equal to 8:", len(password) <= 8)
```
