```python
# 1. create a list of 10 integers and print all even numbers from the list.

numbers = [10, 15, 22, 33, 40, 55, 68, 71, 84, 95]
for num in numbers:
    if num % 2 == 0:
        print("Even numbers from the list are:", num)
```
