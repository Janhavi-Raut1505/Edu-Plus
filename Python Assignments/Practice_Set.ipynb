```python
# 25. Write a program to reverse a number using a while loop.

num = int(input("Enter a number: "))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print("Reversed number =", rev)
```

```python
# 27. Write a program to check whether a number is palindrome or not using a while loop.

num = int(input("Enter a number: "))
original = num
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

if original == reverse:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
# 34. Write a program to check whether a number is Armstrong or not using a while loop.

num = int(input("Enter a number: "))
original = num
sum = 0

while num > 0:
    digit = num % 10
    sum += digit ** 3
    num = num // 10

if sum == original:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# 35. Write a program to print Fibonacci series up to n terms using a while loop.

n = int(input("Enter number of terms: "))

a = 0
b = 1
count = 0

while count < n:
    print(a)
    c = a + b
    a = b
    b = c
    count += 1
```

```python
# 36. Write a program to find the greatest common divisor (GCD) of two numbers using a while loop.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b != 0:
    a, b = b, a % b

print("GCD =", a)
```

```python
# 37. Write a program to find the least common multiple (LCM) of two numbers using a while loop.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

max_num = max(a, b)

while True:
    if max_num % a == 0 and max_num % b == 0:
        print("LCM =", max_num)
        break
    max_num += 1
```

```python
# 39. Write a program to count positive, negative, and zero values from user inputs using a while loop.

positive = negative = zero = 0

while True:
    num = int(input("Enter number (-999 to stop): "))

    if num == -999:
        break

    if num > 0:
        positive += 1
    elif num < 0:
        negative += 1
    else:
        zero += 1

print("Positive =", positive)
print("Negative =", negative)
print("Zero =", zero)
```

```python
# 40. Write a program to create a simple ATM menu using multiple if-else statements.

balance = 10000

print("1. Check Balance")
print("2. Deposit")
print("3. Withdraw")

choice = int(input("Enter your choice: "))

if choice == 1:
    print("Balance =", balance)

elif choice == 2:
    amount = float(input("Enter deposit amount: "))
    balance += amount
    print("Updated Balance =", balance)

elif choice == 3:
    amount = float(input("Enter withdrawal amount: "))
    if amount <= balance:
        balance -= amount
        print("Updated Balance =", balance)
    else:
        print("Insufficient Balance")

else:
    print("Invalid Choice")
```

```python
# 47. Write a program to continuously accept marks until the user enters -1 and then calculate the average.

total = 0
count = 0

while True:
    marks = float(input("Enter marks (-1 to stop): "))

    if marks == -1:
        break

    total += marks
    count += 1

if count > 0:
    average = total / count
    print("Average Marks =", average)
else:
    print("No marks entered")
```

```python
# 48. Write a program to simulate a login system with three attempts using a while loop and if-else conditions.

username = "admin"
password = "1234"

attempts = 0

while attempts < 3:
    user = input("Enter username: ")
    pwd = input("Enter password: ")

    if user == username and pwd == password:
        print("Login Successful")
        break
    else:
        attempts += 1
        print("Invalid Credentials")

if attempts == 3:
    print("Account Locked")
```

```python
# 49. Write a program to check whether a number is perfect or not using a while loop.

num = int(input("Enter a number: "))

sum = 0
i = 1

while i < num:
    if num % i == 0:
        sum += i
    i += 1

if sum == num:
    print("Perfect Number")
else:
    print("Not a Perfect Number")
```
