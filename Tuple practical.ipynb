```python
# 25. Compare the memory size of a list and a tuple containing the same elements using sys.getsizeof().

import sys

# Create a list and tuple with same elements
my_list = [1, 2, 3, 4, 5]
my_tuple = (1, 2, 3, 4, 5)

# Compare memory sizes
list_size = sys.getsizeof(my_list)
tuple_size = sys.getsizeof(my_tuple)

print("Elements:", [1, 2, 3, 4, 5])
print(f"Memory size of List  : {list_size} bytes")
print(f"Memory size of Tuple : {tuple_size} bytes")
print(f"Difference           : {list_size - tuple_size} bytes")
```
